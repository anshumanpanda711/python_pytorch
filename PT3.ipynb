## Using Slices:

```python
torch8 = torch.arange(15)
torch8
```

```python
torch9 = torch8.reshape(5,3)
torch9
```

```python
torch9[:,2]
```

```python
torch9[3,:] # a certain column, or vector
```

How to keep structure?


---


Use a colon at the back:

```python
torch9[:,1:2]
```

```python
torch9[3:,:] # colon after the index results in the 3x1 vectors after and including the index
```

```python
torch9[:3,:] # colon before the index results in the 3x1 vectors before the index
```
