## Using Math Operations:

```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

# Addition:

```python
tensor_a + tensor_b # Shorthand method
```

```python
torch.add(tensor_a, tensor_b) # Longhand method
```

# Subtraction

```python
tensor_b - tensor_a # Shorthand method
```

```python
torch.sub(tensor_b, tensor_a) # Longhand method
```

# Multiplication

```python
tensor_a * tensor_b # Shorthand method
```

```python
torch.mul(tensor_a, tensor_b) # Longhand method
```

# Division

```python
tensor_b / tensor_a # Shorthand method
```

```python
torch.div(tensor_b, tensor_a) # Longhand method
```



---

## Further Operations:
# Modulus (Remainder):

```python
tensor_b % tensor_a # Shorthand method
```

```python
torch.remainder(tensor_b,tensor_a) # Longhand method
```

# Exponentiation:

```python
tensor_a ** tensor_b # Shorthand method
```

```python
torch.pow(tensor_a, tensor_b) # Longhand method
```
